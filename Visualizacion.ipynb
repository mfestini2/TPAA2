```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

models = ["Random Forest", "RNN", "BiRNN"]
capital = [0.5517, 0.7540, 0.8228]
p_ini = [0.6025, 0.7444, 0.8391]
p_fin = [0.3704, 0.5581, 0.7263]

df = pd.DataFrame({
    "Modelo": models,
    "Capitalización": capital,
    "Puntuación Inicial": p_ini,
    "Puntuación Final": p_fin
})

x = np.arange(len(models))
width = 0.25

plt.figure(figsize=(10,6))

plt.bar(x - width, df["Capitalización"], width, label="Capitalización", color="skyblue")
plt.bar(x, df["Puntuación Inicial"], width, label="Puntuación Inicial", color="lightgreen")
plt.bar(x + width, df["Puntuación Final"], width, label="Puntuación Final", color="tomato")

plt.xticks(x, models)
plt.ylabel("F1 Macro")
plt.title("Comparación de modelos por métrica")
plt.ylim(0, None) 
plt.legend()

plt.tight_layout()
plt.show()
```
